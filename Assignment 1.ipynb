## 1 )   Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line. 


```python
# 1st Approch.
for i in range(2000,3201):
    if i%7==0 and i%5!=0:
        print(i,end=",")
```

```python
#2nd Approch.
l1=[]
for i in range(2000,3201):
    if i%7==0 and i%5!=0:
        l1.append(i)
print(l1)
```

```python
#3rd Approch.
l2=[i for i in range(2000,3201) if i%7==0 and i%5 !=0]
print(l2)
```

```python
#4th Approch.
def solution(a,b):
    for i in range(a,b):
        if i%7==0 and i%5!=0:
            yield i
gen=solution(2000,3201)
while True:
    try:
        print(next(gen),end=",")
    except StopIteration:
        break
```

## 2) Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

Senario 1:

```python
#1st Approch.
fname=input("Please Enter your First Name:")
lname=input("Please Enter your last Name:")
print(fname[::-1]+" "+lname[::-1])
```

```python
#2nd Approch.
fname=input("Please Enter your First Name:")
lname=input("Please Enter your last Name:")
def f1(fname):
    rev=''
    for i in range(len(fname),0,-1):
        rev+=fname[i-1]
    return rev
def l1(lname):
    rev=''
    for i in range(len(lname),0,-1):
        rev+=lname[i-1]
    return rev
print(f1(fname)+' '+l1(lname))
```

## 3)  Write a Python program to find the volume of a sphere with diameter 12 cm.

```python
import math
def volume(d):
    r=d/2
    v=(4/3)*(math.pi*(r**3))
    return v
print("The Volume of sphere with radius 12 is: "+ str(volume(12)))
```
